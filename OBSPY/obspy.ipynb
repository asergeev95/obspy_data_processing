```python
from obspy import read
from obspy.core import UTCDateTime
tmp = read("20180505_0600.msd")
tmpD = tmp.copy().detrend()
filtered = tmpD.copy().filter("bandpass", freqmin=0.1, freqmax=30)

summ = filtered.copy() + tmpD.copy() + tmp.copy()  

tmp.plot()
tmpD.plot()
filtered.plot(starttime=UTCDateTime("2018-05-05T06:09:00"), endtime=UTCDateTime("2018-05-05T06:13:00"))
summ.plot()
```
